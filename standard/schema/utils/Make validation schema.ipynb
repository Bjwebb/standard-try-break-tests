```python
import jsonmerge
import json
from jsonschema import validate
```

```python
with open('versioned-release-schema.json', 'rb') as f:
    versioned_release = json.loads(f.read())
```

```python
merger = jsonmerge.Merger(versioned_release)
```

```python
versioned_validation_schema = merger.get_schema()
```

```python
# Check it validates against json schema
with open('tests/json-schema-draft-4.json', 'rb') as f:
    json_schema_4 = json.loads(f.read())
validate(versioned_validation_schema, json_schema_4)
```

```python
print json.dumps(versioned_validation_schema, indent=2)
```

```python
versioned_validation_schema["id"] = "https://raw.githubusercontent.com/open-contracting/standard/master/standard/schema/versioned-release-validation-schema.json"
versioned_validation_schema["$schema"] = "http://json-schema.org/draft-04/schema#"
versioned_validation_schema["title"] = "Schema for a compiled, versioned Open Contracting Release."
```

```python
with open('versioned-release-validation-schema.json', 'wb') as f:
    f.write(json.dumps(versioned_validation_schema, indent=4))
```
